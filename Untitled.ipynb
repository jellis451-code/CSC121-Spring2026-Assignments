```python
def calculate_ticket(age, hour):
    # Determine base price by age
    if age <= 12 or age >= 65:
        price = 8
    else:
        price = 12

    # Apply matinee discount
    if hour < 17:
        price -= 2

    print(f"Your ticket price is ${price:.2f}")


# PROGRAM LOGIC
age = int(input("Enter age: "))
hour = int(input("Enter movie hour (0-23, 24-hour format): "))

calculate_ticket(age, hour)
```
